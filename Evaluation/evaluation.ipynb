```python
import json
import time
from openai import OpenAI
```

```python

# Load input file
with open("llm_eval_input2.json", "r") as f:
    entries = json.load(f)
```

```python
results = []

for i, entry in enumerate(entries):
    try:
        print(f"⏳ Evaluating {i+1}/{len(entries)}")

        response = client.chat.completions.create(
            model="deepseek/deepseek-r1-distill-llama-70b:free",
            messages=[{
                "role": "user",
                "content": entry["prompt"]
            }],
            extra_headers={
                "HTTP-Referer": "https://huggingface.co/datasets/qasper",
                "X-Title": "QasperEval"
            }
        )

        result_text = response.choices[0].message.content

        results.append({
            "paper_id": entry["paper_id"],
            "question": entry["question"],
            "answer_a": entry["answer_a"],
            "answer_b": entry["answer_b"],
            "gold_answer": entry["gold_answer"],
            "llm_judgment": result_text
        })

        time.sleep(2)  # be nice to the API

    except Exception as e:
        print(f"⚠️ Error on entry {i}: {e}")
        print("offending item:", entry)
        continue
```
